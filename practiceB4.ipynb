```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as srs
```

```python
df=pd.read_csv('/content/sample_data/airquality.csv')
```

```python
df.head()
```

```python
labels='Ozone','Solar.R','Temperature','Wind'
```

```python
sizes=[df['Ozone'].mean(),df['Solar.R'].mean(),df['Temp'].mean(),df['Wind'].mean()]
```

```python
colors=['red','yellow','green','brown']
```

```python
explode=(0.5,0,0,0)
```

```python
plt.pie(sizes,explode=explode,labels=labels,colors=colors)
```

```python
h=df.iloc[1:16,1]
v=df.iloc[1:16,3]
```

```python
plt.plot(h,label='Ozone',marker='o',linestyle='dotted')
```

```python
plt.plot(v,label="Ozone",marker='*',linestyle='dashdot')
```

```python
bar=df.nlargest(10,"Ozone")
plt.figure(figsize=(14,12))
srs.barplot(x=bar.Month,y=bar.Ozone)
```
